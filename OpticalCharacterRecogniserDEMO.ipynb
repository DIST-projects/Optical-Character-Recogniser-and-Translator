```
import matplotlib.pyplot as plt   #Used to develop boundary boxes with coordinates assigned using mathematical functions.
import cv2                        #OpenCV used to feed images for demo.
import easyocr                 
from pylab import rcParams          #Command to Run and Configure our matplotlib values
from IPython.display import Image   #Function to display image with respective function overlay applications.
```

```
rcParams['figure.figsize'] = 8, 16
```

```
reader = easyocr.Reader(['hi', 'en'])    #Here we set the module to detect hindi text and English is the most compatible and the language we choose to load model into memory with.
```

```
file_name = "23.jpg"  # Our test image purpose, can feed any image with detectable hindi text subject to model accuracy.
```

```
Image(file_name)    # To display the test image before we apply any function to it
```

```
output = reader.readtext(file_name) #Function reader now applys the OCR using the Devanagari model weights which we have to download manually.
```

##**Here we notice the coordinates of boundary boxes for where the text can be detected along with the text itself.**

```
output           
```

```
cord=output[0][0]
cord1=output[2][0]        #Here we are assigning the gained output coordinates to variables cord and cord1.
```

```
x_min, y_min = [int(min(idx)) for idx in zip(*cord)]
x_max, y_max = [int(max(idx)) for idx in zip(*cord)]
x1_min, y1_min = [int(min(idx)) for idx in zip(*cord1)]
x1_max, y1_max = [int(max(idx)) for idx in zip(*cord1)]   #We assign boundary box coordinates as such.
```

## Now with the coordinate values, we will begin drawing boundary boxes to show the text detected to test our model's accuracy.

```
image = cv2.imread(file_name)
cv2.rectangle(image,(x_min,y_min),(x_max,y_max),(255,0,0),2)
cv2.rectangle(image,(x1_min,y1_min),(x1_max,y1_max),(255,0,0),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))                           
```
